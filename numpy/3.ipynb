```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("========Zomato Sales Data========")
print("\n sales data shape: ", sales_data.shape)
print("\n sample data for 1st 3 restrourants: \n", sales_data[:3])
print("\n sample data for restrourants without id: \n", sales_data[:,1:])
```

```python
# total sales per year of each restaurant
print("\n Total sales per year for all restaurants:")
yearly_total = np.sum(sales_data[:, 1:], axis=0)
print(yearly_total) 

print(np.sum(yearly_total))  # Total sales across all years

# Minimum and maximum sales per restaurant
print("\n Minimum and maximum sales per restaurant:")
print("\n Minimum sales in 2021:", np.min(sales_data[:, 1:], axis=1))   

# max sales per year
print("Maximum sales:", np.max(sales_data[:, 1:], axis=0))

# max sales in 2024
print("Maximum sales in 2024:", np.max(sales_data[:, 1:]))

#average sales per restaurant
print("\n Average sales per restaurant:")
avg_sales  = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales)
```

```python
#cummalative sales per year
print("\n Cumulative sales per year:")
cumulative_sales = np.cumsum(sales_data[:, 1:], axis=1)
print(cumulative_sales)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumulative_sales, axis=0), marker='o', label='Average Cumulative Sales')
plt.title('Average Cumulative Sales Over Years')
plt.xlabel('Year')      
plt.ylabel('Sales (in INR)')
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3,4,5])
vector2 = np.array([6, 7, 8,9,10])
print("\n Vector addition:", vector1 + vector2)  # Element-wise addition
print("Vector subtraction:", vector1 - vector2)  # Element-wise subtraction
print("Vector multiplication:", vector1 * vector2)  # Element-wise multiplication

print("\n Dot product:", np.dot(vector1, vector2))  # Dot product

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("Angle between vectors (in radians):", angle)
```
